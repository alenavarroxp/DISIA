```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from datetime import datetime
import os

# Verificar si hay GPU disponible y configurar TensorFlow para usarla
if tf.config.list_physical_devices('GPU'):
    print("\nGPU disponible. Configurando TensorFlow para usar GPU...")
    physical_devices = tf.config.list_physical_devices('GPU')
    tf.config.experimental.set_memory_growth(physical_devices[0], True)
else:
    print("\nNo se detectó GPU. Se usará CPU.")

# Configuración de rutas para Kaggle
TRAIN_CSV_PATH = '/kaggle/input/dataset/labels/train_data.csv'
TRAIN_IMAGE_DIR = '/kaggle/input/dataset/images'
TEST_CSV_PATH = '/kaggle/input/test-all/labels/test_data.csv'
TEST_IMAGE_DIR = '/kaggle/input/test-all/images'

# Configuración del modelo
IMG_SIZE = (256, 256)  # Tamaño al que se redimensionarán las imágenes
BATCH_SIZE = 32
EPOCHS = 50
INIT_LR = 0.0001

# 1. Cargar y verificar los datos de entrenamiento
print("\nCargando datos de entrenamiento...")
train_df = pd.read_csv(TRAIN_CSV_PATH)
print(f"Total de muestras de entrenamiento: {len(train_df)}")

# Verificar que las imágenes de entrenamiento existen
sample_train_image_path = os.path.join(TRAIN_IMAGE_DIR, f"{train_df.iloc[0]['ID']}.jpg")
if not os.path.exists(sample_train_image_path):
    raise FileNotFoundError(f"No se encontró la imagen de entrenamiento en {sample_train_image_path}")

# 2. Cargar y verificar los datos de prueba (si es necesario)
print("\nCargando datos de prueba...")
test_df = pd.read_csv(TEST_CSV_PATH)
print(f"Total de muestras de prueba: {len(test_df)}")

sample_test_image_path = os.path.join(TEST_IMAGE_DIR, f"{test_df.iloc[0]['ID']}.jpg")
if not os.path.exists(sample_test_image_path):
    raise FileNotFoundError(f"No se encontró la imagen de prueba en {sample_test_image_path}")

# 3. Dividir los datos de entrenamiento (estratificado para mantener proporción de clases)
print("\nDividiendo datos de entrenamiento...")
train_df, val_df = train_test_split(
    train_df, 
    test_size=0.2, 
    random_state=42, 
    stratify=train_df[['transitable', 'Inundado']]
)
print(f"Muestras de entrenamiento: {len(train_df)}")
print(f"Muestras de validación: {len(val_df)}")

# 4. Generador de datos optimizado para GPU
class BWImageGenerator(tf.keras.utils.Sequence):
    def __init__(self, df, img_dir, batch_size=32, shuffle=True, augment=False):
        self.df = df.reset_index(drop=True)
        self.img_dir = img_dir
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.augment = augment
        self.on_epoch_end()
        
        # Pre-calcular los pesos para evitar problemas de índice
        self._precompute_sample_weights()
        
        # Configuración de aumento de datos
        self.augmenter = ImageDataGenerator(
            rotation_range=20,
            width_shift_range=0.1,
            height_shift_range=0.1,
            zoom_range=0.1,
            horizontal_flip=True,
            fill_mode='constant',
            cval=0
        )
    
    def _precompute_sample_weights(self):
        # Calcular weights para transitable
        transitable_classes = np.unique(self.df['transitable'])
        transitable_weights = class_weight.compute_class_weight(
            'balanced',
            classes=transitable_classes,
            y=self.df['transitable']
        )
        self.transitable_weights = np.array([transitable_weights[0] if x == 0 else transitable_weights[1] for x in self.df['transitable']])
        
        # Calcular weights para inundado
        inundado_classes = np.unique(self.df['Inundado'])
        inundado_weights = class_weight.compute_class_weight(
            'balanced',
            classes=inundado_classes,
            y=self.df['Inundado']
        )
        self.inundado_weights = np.array([inundado_weights[0] if x == 0 else inundado_weights[1] for x in self.df['Inundado']])
    
    def __len__(self):
        return int(np.ceil(len(self.df) / self.batch_size))
    
    def __getitem__(self, index):
        start_idx = index * self.batch_size
        end_idx = min((index + 1) * self.batch_size, len(self.df))
        batch_indices = self.indices[start_idx:end_idx]
        current_batch_size = len(batch_indices)
        
        # Pre-asignar memoria para mejor rendimiento con GPU
        X = np.empty((current_batch_size, *IMG_SIZE, 1), dtype=np.float32)
        y_transitable = np.empty((current_batch_size, 1), dtype=np.float32)
        y_inundado = np.empty((current_batch_size, 1), dtype=np.float32)
        sw_transitable = np.empty((current_batch_size, 1), dtype=np.float32)
        sw_inundado = np.empty((current_batch_size, 1), dtype=np.float32)
        
        for i, idx in enumerate(batch_indices):
            row = self.df.iloc[idx]
            img_path = os.path.join(self.img_dir, f"{row['ID']}.jpg")
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            
            if img is None:
                raise ValueError(f"No se pudo cargar la imagen: {img_path}")
                
            img = cv2.resize(img, IMG_SIZE)
            img = img.reshape(*IMG_SIZE, 1)
            
            if self.augment:
                img = self.augmenter.random_transform(img)
                
            X[i,] = img / 255.0  # Normalización
            
            y_transitable[i,] = row['transitable']
            y_inundado[i,] = row['Inundado']
            sw_transitable[i,] = self.transitable_weights[idx]
            sw_inundado[i,] = self.inundado_weights[idx]
            
        # Convertir a tensores para mejor compatibilidad con GPU
        X = tf.convert_to_tensor(X, dtype=tf.float32)
        y_transitable = tf.convert_to_tensor(y_transitable, dtype=tf.float32)
        y_inundado = tf.convert_to_tensor(y_inundado, dtype=tf.float32)
        sw_transitable = tf.convert_to_tensor(sw_transitable, dtype=tf.float32)
        sw_inundado = tf.convert_to_tensor(sw_inundado, dtype=tf.float32)
            
        return X, {'transitable': y_transitable, 'inundado': y_inundado}, {'transitable': sw_transitable, 'inundado': sw_inundado}
    
    def on_epoch_end(self):
        self.indices = np.arange(len(self.df))
        if self.shuffle:
            np.random.shuffle(self.indices)

# 5. Arquitectura del modelo optimizada para GPU
def create_model():
    inputs = Input(shape=(*IMG_SIZE, 1), batch_size=BATCH_SIZE)
    
    # Bloques convolucionales optimizados
    x = Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.3)(x)
    
    x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.4)(x)
    
    x = Conv2D(256, (3, 3), activation='relu', padding='same')(x)
    x = BatchNormalization()(x)
    x = GlobalAveragePooling2D()(x)
    
    # Capas densas
    x = Dense(512, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    
    # Ramas de salida
    transitable_output = Dense(1, activation='sigmoid', name='transitable')(x)
    inundado_output = Dense(1, activation='sigmoid', name='inundado')(x)
    
    model = Model(inputs=inputs, outputs=[transitable_output, inundado_output])
    
    return model

# 6. Compilar modelo para GPU
def compile_model(model):
    model.compile(
        optimizer=Adam(learning_rate=INIT_LR),
        loss={
            'transitable': 'binary_crossentropy',
            'inundado': 'binary_crossentropy'
        },
        loss_weights={
            'transitable': 1.0,
            'inundado': 1.0
        },
        metrics={
            'transitable': ['accuracy', tf.keras.metrics.Precision(name='precision'), tf.keras.metrics.Recall(name='recall')],
            'inundado': ['accuracy', tf.keras.metrics.Precision(name='precision'), tf.keras.metrics.Recall(name='recall')]
        }
    )
    return model

# 7. Callbacks para Kaggle con GPU
def get_callbacks():
    return [
        tf.keras.callbacks.EarlyStopping(
            patience=15,
            monitor='val_loss',
            restore_best_weights=True,
            min_delta=0.001,
            verbose=1
        ),
        tf.keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=5,
            min_lr=1e-7,
            verbose=1
        ),
        tf.keras.callbacks.ModelCheckpoint(
            'best_model.keras',
            save_best_only=True,
            monitor='val_loss',
            mode='min',
            verbose=1
        ),
        tf.keras.callbacks.CSVLogger('training_log.csv'),
        tf.keras.callbacks.TensorBoard(log_dir='./logs')
    ]

# 8. Entrenamiento del modelo con GPU
def train_model():
    # Crear generadores
    train_gen = BWImageGenerator(
        train_df, 
        img_dir=TRAIN_IMAGE_DIR, 
        augment=True,
        shuffle=True,
        batch_size=BATCH_SIZE
    )
    
    val_gen = BWImageGenerator(
        val_df, 
        img_dir=TRAIN_IMAGE_DIR, 
        augment=False,
        shuffle=False,
        batch_size=BATCH_SIZE
    )
    
    # Crear y compilar modelo
    model = create_model()
    model = compile_model(model)
    print("\nResumen del modelo optimizado para GPU:")
    model.summary()
    
    # Entrenamiento
    print("\nIniciando entrenamiento con GPU...")
    history = model.fit(
        train_gen,
        validation_data=val_gen,
        epochs=EPOCHS,
        callbacks=get_callbacks(),
        verbose=1
    )
    
    return model, history

# 9. Evaluación del modelo con datos de prueba
def evaluate_on_test(model):
    test_gen = BWImageGenerator(
        test_df,
        img_dir=TEST_IMAGE_DIR,
        augment=False,
        shuffle=False,
        batch_size=BATCH_SIZE
    )
    
    print("\nEvaluando en conjunto de prueba...")
    results = model.evaluate(test_gen, verbose=1)
    
    print("\n=== Resultados en Prueba ===")
    print(f"Pérdida total: {results[0]:.4f}")
    print(f"Pérdida transitable: {results[1]:.4f}")
    print(f"Pérdida inundado: {results[2]:.4f}")
    print(f"Accuracy transitable: {results[3]:.4f}")
    print(f"Precisión transitable: {results[4]:.4f}")
    print(f"Recall transitable: {results[5]:.4f}")
    print(f"Accuracy inundado: {results[6]:.4f}")
    print(f"Precisión inundado: {results[7]:.4f}")
    print(f"Recall inundado: {results[8]:.4f}")

# 10. Visualización de resultados
def plot_results(history):
    plt.figure(figsize=(18, 12))
    
    # Pérdida
    plt.subplot(2, 2, 1)
    plt.plot(history.history['loss'], label='Entrenamiento')
    plt.plot(history.history['val_loss'], label='Validación')
    plt.title('Evolución de la Pérdida')
    plt.ylabel('Pérdida')
    plt.xlabel('Época')
    plt.legend()
    
    # Accuracy Transitable
    plt.subplot(2, 2, 2)
    plt.plot(history.history['transitable_accuracy'], label='Entrenamiento')
    plt.plot(history.history['val_transitable_accuracy'], label='Validación')
    plt.title('Accuracy - Transitable')
    plt.ylabel('Accuracy')
    plt.xlabel('Época')
    plt.legend()
    
    # Accuracy Inundado
    plt.subplot(2, 2, 3)
    plt.plot(history.history['inundado_accuracy'], label='Entrenamiento')
    plt.plot(history.history['val_inundado_accuracy'], label='Validación')
    plt.title('Accuracy - Inundado')
    plt.ylabel('Accuracy')
    plt.xlabel('Época')
    plt.legend()
    
    # Pérdida por tarea
    plt.subplot(2, 2, 4)
    plt.plot(history.history['transitable_loss'], label='Transitable (Entren.)')
    plt.plot(history.history['val_transitable_loss'], label='Transitable (Val.)')
    plt.plot(history.history['inundado_loss'], label='Inundado (Entren.)')
    plt.plot(history.history['val_inundado_loss'], label='Inundado (Val.)')
    plt.title('Pérdida por Tarea')
    plt.ylabel('Pérdida')
    plt.xlabel('Época')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# 11. Ejecución principal
if __name__ == "__main__":
    print("\n=== Configuración Inicial ===")
    print(f"TensorFlow está usando GPU: {len(tf.config.list_physical_devices('GPU')) > 0}")
    print(f"Ruta imágenes entrenamiento: {TRAIN_IMAGE_DIR}")
    print(f"Ruta imágenes prueba: {TEST_IMAGE_DIR}")
    
    try:
        # Entrenar modelo
        trained_model, training_history = train_model()
        
        # Evaluar en validación
        val_gen = BWImageGenerator(val_df, img_dir=TRAIN_IMAGE_DIR, shuffle=False, batch_size=BATCH_SIZE)
        print("\nEvaluando en conjunto de validación...")
        trained_model.evaluate(val_gen, verbose=1)
        
        # Evaluar en prueba
        evaluate_on_test(trained_model)
        
        # Visualizar resultados
        plot_results(training_history)
        
        # Guardar modelo final
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

        trained_model.save(f'final_model_CCN{timestamp}.keras')
 
        
    except Exception as e:
        print(f"\nError durante la ejecución: {e}")
        raise
```
