```python
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Input, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import (ReduceLROnPlateau, EarlyStopping, 
                                      ModelCheckpoint, CSVLogger, TensorBoard)
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
import numpy as np
import pandas as pd
import cv2
import os
import matplotlib.pyplot as plt
from datetime import datetime

print("\n=== Configuración Inicial ===")
physical_devices = tf.config.list_physical_devices('GPU')
if physical_devices:
    tf.config.experimental.set_memory_growth(physical_devices[0], True)
    print("GPU configurada con memory_growth")
else:
    print("No se detectó GPU, usando CPU")

IMG_SIZE = (224, 224)
BATCH_SIZE = 32
EPOCHS = 50
INIT_LR = 1e-4

TRAIN_CSV = '/kaggle/input/dataset/labels/train_data.csv'
TRAIN_IMG_DIR = '/kaggle/input/dataset/images'
TEST_CSV = '/kaggle/input/test-all/labels/test_data.csv'
TEST_IMG_DIR = '/kaggle/input/test-all/images'

print("\nCargando datos...")
train_df = pd.read_csv(TRAIN_CSV)
test_df = pd.read_csv(TEST_CSV)

sample_train_img = os.path.join(TRAIN_IMG_DIR, f"{train_df.iloc[0]['ID']}.jpg")
sample_test_img = os.path.join(TEST_IMG_DIR, f"{test_df.iloc[0]['ID']}.jpg")

if not os.path.exists(sample_train_img):
    raise FileNotFoundError(f"No se encontró imagen: {sample_train_img}")
if not os.path.exists(sample_test_img):
    raise FileNotFoundError(f"No se encontró imagen: {sample_test_img}")

print("\nDividiendo datos...")
train_df, val_df = train_test_split(
    train_df,
    test_size=0.2,
    random_state=42,
    stratify=train_df[['transitable', 'Inundado']]
)
print(f"Entrenamiento: {len(train_df)}, Validación: {len(val_df)}, Prueba: {len(test_df)}")

class MultiTargetGenerator(tf.keras.utils.Sequence):
    def __init__(self, df, img_dir, batch_size=32, shuffle=True, augment=False):
        self.df = df.reset_index(drop=True)
        self.img_dir = img_dir
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.augment = augment
        self.on_epoch_end()
        self._precompute_class_weights()
        self.augmenter = ImageDataGenerator(
            rotation_range=30,
            width_shift_range=0.2,
            height_shift_range=0.2,
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True,
            vertical_flip=True,
            brightness_range=[0.8, 1.2],
            fill_mode='constant',
            cval=0
        )

    def _precompute_class_weights(self):
        t_classes = np.unique(self.df['transitable'])
        t_weights = class_weight.compute_class_weight('balanced', classes=t_classes, y=self.df['transitable'])
        self.transitable_weights = np.array([t_weights[0] if x == 0 else t_weights[1] for x in self.df['transitable']])

        i_classes = np.unique(self.df['Inundado'])
        i_weights = class_weight.compute_class_weight('balanced', classes=i_classes, y=self.df['Inundado'])
        self.inundado_weights = np.array([i_weights[0] if x == 0 else i_weights[1] for x in self.df['Inundado']])

    def __len__(self):
        return int(np.ceil(len(self.df) / self.batch_size))

    def __getitem__(self, index):
        idxs = self.indices[index*self.batch_size:(index+1)*self.batch_size]
        X = np.empty((len(idxs), *IMG_SIZE, 3), dtype=np.float32)
        y_t = np.empty((len(idxs), 1), dtype=np.float32)
        y_i = np.empty((len(idxs), 1), dtype=np.float32)
        sw_t = np.empty((len(idxs), 1), dtype=np.float32)
        sw_i = np.empty((len(idxs), 1), dtype=np.float32)

        for i, idx in enumerate(idxs):
            row = self.df.iloc[idx]
            img_path = os.path.join(self.img_dir, f"{row['ID']}.jpg")
            img = cv2.imread(img_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = cv2.resize(img, IMG_SIZE)
            if self.augment:
                img = self.augmenter.random_transform(img)
            X[i] = img / 255.0
            y_t[i] = row['transitable']
            y_i[i] = row['Inundado']
            sw_t[i] = self.transitable_weights[idx]
            sw_i[i] = self.inundado_weights[idx]

        return X, {'transitable': y_t, 'inundado': y_i}, {'transitable': sw_t, 'inundado': sw_i}

    def on_epoch_end(self):
        self.indices = np.arange(len(self.df))
        if self.shuffle:
            np.random.shuffle(self.indices)

def build_vgg16_model():
    base_model = VGG16(include_top=False, weights='imagenet', input_shape=(*IMG_SIZE, 3))
    for layer in base_model.layers[:15]:
        layer.trainable = False

    inputs = Input(shape=(*IMG_SIZE, 3), batch_size=BATCH_SIZE)
    x = base_model(inputs)
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    x = Dense(512, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)

    t = Dense(256, activation='relu')(x)
    t = BatchNormalization()(t)
    t = Dropout(0.3)(t)
    t_out = Dense(1, activation='sigmoid', name='transitable')(t)

    i = Dense(256, activation='relu')(x)
    i = BatchNormalization()(i)
    i = Dropout(0.3)(i)
    i_out = Dense(1, activation='sigmoid', name='inundado')(i)

    return Model(inputs, [t_out, i_out])

def compile_model(model):
    model.compile(
        optimizer=Adam(INIT_LR),
        loss={'transitable': 'binary_crossentropy', 'inundado': 'binary_crossentropy'},
        loss_weights={'transitable': 1.0, 'inundado': 1.0},
        metrics={
            'transitable': ['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), tf.keras.metrics.AUC()],
            'inundado': ['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), tf.keras.metrics.AUC()]
        }
    )
    return model

def get_callbacks():
    ts = datetime.now().strftime("%Y%m%d_%H%M%S")
    return [
        EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, min_lr=1e-6),
        ModelCheckpoint(f'best_model_vgg16_{ts}.keras', save_best_only=True),
        CSVLogger(f'training_log_vgg16_{ts}.csv'),
        TensorBoard(log_dir=f'./logs/vgg16_{ts}')
    ]

def plot_results(history):
    keys = history.history.keys()
    print("Métricas registradas:", keys)

    plt.figure(figsize=(18, 12))

    # Loss total
    plt.subplot(2, 2, 1)
    plt.plot(history.history['loss'], label='Train')
    plt.plot(history.history['val_loss'], label='Val')
    plt.title('Total Loss')
    plt.legend()

    # Accuracy
    plt.subplot(2, 2, 2)
    if 'transitable_accuracy' in keys:
        plt.plot(history.history['transitable_accuracy'], label='Trans. Train')
    if 'val_transitable_accuracy' in keys:
        plt.plot(history.history['val_transitable_accuracy'], label='Trans. Val')
    if 'inundado_accuracy' in keys:
        plt.plot(history.history['inundado_accuracy'], label='Inund. Train')
    if 'val_inundado_accuracy' in keys:
        plt.plot(history.history['val_inundado_accuracy'], label='Inund. Val')
    plt.title('Accuracy')
    plt.legend()

    # AUC
    plt.subplot(2, 2, 3)
    if 'transitable_auc' in keys:
        plt.plot(history.history['transitable_auc'], label='Trans. AUC Train')
    if 'val_transitable_auc' in keys:
        plt.plot(history.history['val_transitable_auc'], label='Trans. AUC Val')
    if 'inundado_auc' in keys:
        plt.plot(history.history['inundado_auc'], label='Inund. AUC Train')
    if 'val_inundado_auc' in keys:
        plt.plot(history.history['val_inundado_auc'], label='Inund. AUC Val')
    plt.title('AUC')
    plt.legend()

    # Loss por tarea
    plt.subplot(2, 2, 4)
    if 'transitable_loss' in keys:
        plt.plot(history.history['transitable_loss'], label='Trans. Loss Train')
    if 'val_transitable_loss' in keys:
        plt.plot(history.history['val_transitable_loss'], label='Trans. Loss Val')
    if 'inundado_loss' in keys:
        plt.plot(history.history['inundado_loss'], label='Inund. Loss Train')
    if 'val_inundado_loss' in keys:
        plt.plot(history.history['val_inundado_loss'], label='Inund. Loss Val')
    plt.title('Loss por tarea')
    plt.legend()

    plt.tight_layout()
    plt.show()

def evaluate_model(model, test_gen):
    print("\n=== Evaluación ===")
    results = model.evaluate(test_gen)
    for name, val in zip(model.metrics_names, results):
        print(f"{name}: {val:.4f}")

# MAIN
if __name__ == "__main__":
    try:
        train_gen = MultiTargetGenerator(train_df, TRAIN_IMG_DIR, BATCH_SIZE, shuffle=True, augment=True)
        val_gen = MultiTargetGenerator(val_df, TRAIN_IMG_DIR, BATCH_SIZE, shuffle=False, augment=False)
        test_gen = MultiTargetGenerator(test_df, TEST_IMG_DIR, BATCH_SIZE, shuffle=False, augment=False)

        model = build_vgg16_model()
        model = compile_model(model)
        model.summary()

        history = model.fit(
            train_gen,
            validation_data=val_gen,
            epochs=EPOCHS,
            callbacks=get_callbacks(),
            verbose=1
        )

        plot_results(history)
        evaluate_model(model, test_gen)

        ts = datetime.now().strftime("%Y%m%d_%H%M%S")
        model.save(f'final_model_vgg16_{ts}.keras')
        print(f"\n✅ Modelo final guardado como 'final_model_vgg16_{ts}.keras'")

    except Exception as e:
        print(f"❌ Error: {e}")
        raise
```
